```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
```

```python
df = data.drop(["Row ID","Order ID","Customer ID","Product ID"], axis = 1)
data2 = data[["Row ID","Order ID","Customer ID","Product ID"]]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[df["Postal Code"].isnull()]
```

```python
#df[df["City"] == "Burlington"].fillna("05001.0")
df["Postal Code"] = df["Postal Code"].fillna(05001.0)
```

```python
#Ay Sütunu
df["Month"] = df["Order Date"].str[3:5]
```

```python
df
```

```python
df["Month"] = df["Month"].astype("int32")
```

```python
df
```

```python
d = {
    1: "Ocak",
    2: "Şubat",
    3: "Mart",
    4: "Nisan",
    5: "Mayıs",
    6: "Haziran",
    7: "Temmuz",
    8: "Ağustos",
    9: "Eylül",
    10: "Ekim",
    11: "Kasım",
    12: "Aralık"
}

df["Month"] = [d[i] for i in df["Month"]]
```

```python
df
```

```python
df["Order Date"] = pd.to_datetime(df["OrderDate"], infer_datetime_format=True)
```

```python
df
```

```python
df["ShipDate"] = pd.to_datetime(df["Ship Date"])
```

```python
df
```

```python
df = df.drop(["OrderDate","ShipDate"],axis=1)
df["Year"] = df["Order Date"].dt.year  
```

```python
df
```

```python
df["Year"].value_counts()
```

```python
months_df = df.groupby("Month").sum()
months_df
```

```python
months_df = months_df[["Sales"]]
months_df
```

```python
months_df = months_df.reset_index()
```

```python
months_df
```

```python
plt.bar(months_df["Month"],months_df["Sales"])
plt.xlabel("Satış Miktarı")
plt.ylabel("Aylar")
plt.xticks(fontsize=12,rotation=90)
plt.yticks(fontsize=12)
plt.show()
```

```python
year_df = df.groupby("Year").sum()
year_df
```

```python
year_df = year_df[["Sales"]]
year_df
```

```python
year_df = year_df.reset_index()
```

```python
year_df
```

```python
plt.bar(year_df["Year"],year_df["Sales"])
plt.xlabel("Yıllar")
plt.ylabel("Satış Miktarı")
```

```python
Customers = df.groupby("Customer Name").sum().sort_values("Sales",ascending=False)
Customers.drop(["Postal Code","Year"],axis=1,inplace=True)
Customers
```

```python
Customers.reset_index(inplace=True)
```

```python
Customers
```

```python
Customers = Customers.head(20)
```

```python
plt.bar(Customers["Customer Name"],Customers["Sales"])
plt.xlabel("Kişi")
plt.ylabel("Satış")
plt.xticks(fontsize=12,rotation=90)
plt.yticks(fontsize=12)
plt.show()
```

```python
category = df.groupby("Category").sum().sort_values("Sales",ascending=False)
category.reset_index(inplace=True)
category
```

```python
category = category[["Category","Sales"]]
category
```
